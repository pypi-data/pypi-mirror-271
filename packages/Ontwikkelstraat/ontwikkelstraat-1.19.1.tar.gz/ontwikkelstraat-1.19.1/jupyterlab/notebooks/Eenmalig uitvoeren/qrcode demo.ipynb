```python
import qrcode 
import qrcode.image.svg
href="https://delen.meteddie.nl/item/DaHF6UVyTaavTxTCd8kMlw?utm_source=narrowcasting-lef"
img = qrcode.make(
    href, 
    image_factory=qrcode.image.svg.SvgPathImage
)
svg_html = img.to_string().decode()
display(HTML(svg_html))

# Nu met embedded PNG: 
qr = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_M)
import qrcode.image.styledpil
qr.add_data(href)
img = qr.make_image(
    image_factory=qrcode.image.styledpil.StyledPilImage, 
    embeded_image_path="eddie-icon.png"
)
display(img)
```
