```python
from skyrim import Skyrim
```

```python
# to see all the available models
Skyrim.list_available_models()
```

```python
# lets initialize a pangu model using Skyrim API
# this will download the model if not already cached
model = Skyrim("pangu")
```

```python
# available functions
print([attr for attr in dir(model) if not attr.startswith("__")])
```

```python
# pred: GlobalPrediction
#   holds final prediction and its metadata
#   GlobalPrediction is thin wrapper around xarray.Dataset for data manipulation
# output paths: list[str]
#   list of paths where the output is saved is save=True

pred, output_paths = model.predict(
    date="20180101", time="0000", lead_time=24, save=True
)
```
